###Ερώτημα 2: Υπολογισμός παραγώγων




**2.1** $ $ Έστω η συνάρτηση 


$\ \ \ \ \ \ \ \ \ \ \ \ \ \ f(\textbf{x})=\textbf{x}^{T}A\textbf{x}+\textbf{b}^{T}\textbf{x}$ ,



$\ \ \ \ \ \ $ όπου $A$  συμμετρικός πίνακας ($A=A^{T}$) διάστασης $d\times d$, και $\textbf{b}\in\mathbb{R}^{d}$.     

$\ \ \ \ \ \ $ Να βρεθεί η παράγωγος $\textbf{x}\in\mathbb{R}$ στο $f(\textbf{x})$.


**Λύση** $ $ $ $  Έχουμε τη συνάρτηση $f(\textbf{x})=\textbf{x}^{T}A\textbf{x}+\textbf{b}^{T}\textbf{x}$

Όπου $A$    $d\times d$ συμμετρικός πίνακας ($A=A^{T}$), και $\textbf{b}\in\mathbb{R}^{d}$. θέλουμε την παράγωγο $\textbf{x}\in\mathbb{R}$ στο $f(\textbf{x})$. Άρα $f:\mathbb{R}^{d}\rightarrow\mathbb{R}$,  

$ $

$$\nabla f({\textbf{x}})= \begin{pmatrix}
    \ \frac{\partial f(\textbf{x})}{\partial x_{1}} \ \ \\
    \ \frac{\partial f(\textbf{x})}{\partial x_{2}}    \\
    \vdots \\
    \ \frac{\partial f(\textbf{x})}{\partial x_{d}}    \\
\end{pmatrix}
\
\in \mathbb{R}^{d}
$$

$ $
 
όπου $$ \vec{x} = (x_{1}, x_{2}, \dots , x_{d}) \doteq 
\begin{pmatrix}
    \ x_{1} \ \\
    \ x_{2}    \\
    \vdots \\
    \ x_{d}    \\
\end{pmatrix}
 $$ 

$ $

-----------------------

$ $

Υπολογίζουμε για κάποιο $i \in \mathbb{R}^{d} \  $ την $  \ \frac{\partial f(\textbf{x})}{\partial x_{i}}$, έχουμε
 
 $ $ 

$$\frac{\partial f( \vec{x} )}{\partial x_{i}} = \frac{\partial f(x_{1},x_{2}, \dots, x_{i}, \dots, x_{d}) }{\partial x_{i}} = \frac{\partial ( \vec{x}^{T}A\vec{x}+\textbf{b}^{T}\vec{x}) }{\partial x_{i}} = \frac{\partial ( \vec{x}^{T}A\vec{x}) }{\partial x_{i}}+ \frac{\partial ( \textbf{b}^{T}\vec{x}) }{\partial x_{i}} \ \ \ \ \ \ \ \ \bf (1) $$

$ $

$ \vec{x} \in \mathbb{R}^{d}$, όμως $f(\vec{x}) \in \mathbb{R}$, δηλαδή  $f(\vec{x})$ είναι βαθμωτό μέγεθος ενώ το όρισμα της (το $ \vec{x}$) είναι διάνυσμα $(x_{1}, x_{2}, \dots , x_{d})$.

$ $

Πρώτος όρος, 

$ $

$$\frac{\partial ( \vec{x}^{T}A\vec{x}) }{\partial x_{i}} \ \ = \ \ \frac{\partial  \vec{x}^{T} }{\partial x_{i}}Α\vec{x} \ + \ \vec{x}^{T}A\frac{\partial \vec{x} }{\partial x_{i}} \ \ = \  \ \frac{∂ (x_{1}, x_{2}, \dots , x_{d})}{\partial x_{i}}Α\vec{x} \ + \ \vec{x}^{T}A \frac{∂ \begin{pmatrix}
     x_{1} \ \\
    \ x_{2}    \\
    \vdots \\
    \ x_{d}    \\
\end{pmatrix}}{\partial x_{i}} \ \ = \ \ $$

$ $

$$ = \ \ \left ( \frac{∂x_{1}}{\partial x_{i}} \ \frac{∂x_{2}}{\partial x_{i}} \ \dots \ \frac{∂x_{i}}{\partial x_{i}} \ \dots \ \frac{∂x_{d}}{\partial x_{i}} \right)Α\vec{x} \ + \ \vec{x}^{T}A \begin{pmatrix}
     \frac{∂x_{1}}{\partial x_{i}} \ \\
    \ \frac{∂x_{2}}{\partial x_{i}}    \\
    \vdots \\
    \ \frac{∂x_{i}}{\partial x_{i}}    \\
    \ \vdots \\
    \frac{∂x_{d}}{\partial x_{i}} \\
\end{pmatrix}
$$

$ $

όμως 

$ $

$$ \frac{∂x_{1}}{\partial x_{i}}  \ = \ 0 \ \ \ \ \ \frac{∂x_{2}}{\partial x_{i}}\ = \ 0  \ \ \ \ \ \dots \ \ \ \ \  \frac{∂x_{i}}{\partial x_{i}}\ = \ 1 \ \ \ \ \ \dots \ \ \ \ \ \frac{∂x_{d}}{\partial x_{i}} \ = \ 0  $$

$ $

Άρα

$$ \frac{\partial ( \vec{x}^{T}A\vec{x}) }{\partial x_{i}} \ \ = \ \ (0 \ 0 \ \dots \ 1 \ \dots \ 0)Α\vec{x} \ + \ \vec{x}^{T}A
 \begin{pmatrix}
     0  \\
    \ 0    \\
    \vdots \\
    \ 1    \\
    \ \vdots \\
    \ 0 \\
\end{pmatrix} \ \ = \ \ V_i^T Α\vec{x} \ + \ \vec{x}^{T}A V_i \ \ \ \ \ \ \ \ \bf (2)$$

Όπου, $V_i$ αναπαριστάται από, 

$ $

$$\textbf{V}_{i}= \begin{pmatrix}
     0 \ \\
    \ 0    \\

    \ \vdots \\
    \ 0 \\
\end{pmatrix} 
\
$$

$ $

Επιπλέον $ \ \vec{x}^{T}A V_i \ , \ $ είναι βαθμωτό μέγεθος, άρα ισχύει ότι

$$ \vec{x}^{T}A V_i \ \ = \ \ (\vec{x}^{T}A V_i)^{Τ} \ \ = \ \ V_i^{T}A^{T}\vec{x} $$

και αφού ο πίνακας Α είναι συμμετρικός, έχουμε
$$ \vec{x}^{T}A V_i \ \ = \ \ V_i^{T}A^{T}\vec{x} \ \ = \ \ V_i^{T}A\vec{x} $$

$ $

Επομένως η (2) γράφεται
$$ \frac{\partial ( \vec{x}^{T}A\vec{x}) }{\partial x_{i}} \ \ = \ \ 2 \cdot V_i^{T}A\vec{x} \ \ = \ \ 2 \cdot (0 \dots 1 \dots 0)

###Ερώτημα 3: Gradient descent

Ο αλγόριθμος $\ \it Gradient Descent \ $, για την ελαχιστοποίηση συναρτήσεων 2 μεταβλητών είναι παρόμοιος με αυτόν για μια μεταβλητή. Για τις ζητούμενες συναρτήσεις $f_{1}$ και $f_{2}$, οι αντίστοιχες μερικές παραγώγους τους, για τον υπολογισμό του gradient $\nabla_x f({\textbf{x}}) $ είναι οι εξής:

$ $
$$ f_{1}: \ \ \ \ \nabla f(x_{1}, x_{2}) \ \ = \ \ \frac{∂f_1}{x_{1}} \ + \ \frac{∂f_1}{x_{2}} \ \ = \ \ 2(x_{1} \ - \ 2) \ + \ 2(x_{2} \ - \ 3)$$

$ $
$$ \text{και} $$

$ $
$$ 
f_{2}: \ \ \ \ \nabla f(x_{1}, x_{2}) \ \ = \ \ \frac{∂f_2}{x_{1}} \ + \ \frac{∂f_2}{x_{2}} \ \ = \ \ 40((x_1 \ + \ 3) \ - \ (x_2 \ - \ 3)^2) \ + \ (-2)(4 \ - \ x_2) \ + \ \hspace{70mm} + \ 40((x_1 \ + \ 3) \ - \ (x_2 \ - \ 3)^2)(-2(x2-3)))
$$

$ $

Παρακάτω υλοποιούμε τον αλγόριθμο για 2 μεταβλητές, ο οποίος είναι μία τροποποίηση του αλγορίθμου του φροντιστηρίου.

$$ \bullet \ \ f_{1}(x_{1},x_{2}) = (x_{1}-2)^{2} + (x_{2}-3)^{2} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  $$

```python
import matplotlib.pyplot as plt

def df(x1,x2):
  return (2*(x1-2), 2*(x2-3))

curX = (0.0, 0.0)   # initialize X
rate = 0.5      # Learning rate
precision = 0.0001    #This tells us when to stop the algorithm
previous_step_size = 1 #
maxIters = 10     # maximum number of iterations
iters = 0         #iteration counter

# save info for the graph
sols = [(curX[0] - 2)**2 + (curX[1] - 3)**2]
inters = [0]

print("Iteration",0,"\nX value is",curX)    #Print iterations

for aIter in range(maxIters):
  prevX=curX
  a1, a2 = df(curX[0],curX[1])
  b1 = rate*a1
  b2 = rate*a2
  c1 = curX[0] - b1
  c2 = curX[1] - b2
  curX = (c1,c2)    # new X
  sols.append((curX[0] - 2)**2 + (curX[1] - 3)**2)
  for i in range(2):
    diff=abs(curX[i]-prevX[i])
    if diff<=precision:
      print("\nThe local minimum occurs at", curX)
      break
  print("Iteration",aIter+1,"\nX value is",curX) #Print iterations
  inters.append(aIter+1)
  
# Graph
plt.figure(figsize=(10, 3)) 
plt.plot(inters, sols)
plt.ylabel('f(x1,x2)')
plt.xlabel('Iteration')
plt.title("Gradient Descent")
plt.show()

```

$$ \bullet \ \   f_{2}(x_{1},x_{2}) = (1-(x_{2}-3))^{2} + 20((x_{1}+3)-(x_{2}-3)^{2})  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ $$

```python
import matplotlib.pyplot as plt

def df(x1,x2):
  return (40*((x1+3)-(x2-3)**2),-2*(4-x2)+40*((x1+3)-(x2-3)**2)*(-2*(x2-3)))

curX = (0.0, 0.0)   # initialize X
rate = 0.5   # Learning rate
precision = 0.0001    #This tells us when to stop the algorithm
previous_step_size = 1 #
maxIters = 10     # maximum number of iterations
iters = 0         #iteration counter

# save info for the graph
sols = [(1-(curX[1]-3))**2+20*((curX[0]+3)-(curX[1]-3)**2)**2]
inters = [0]

print("Iteration",0,"\nX value is",curX)    #Print iterations

for aIter in range(maxIters):
  prevX=curX
  a1, a2 = df(curX[0],curX[1])
  b1 = rate*a1
  b2 = rate*a2
  c1 = curX[0] - b1
  c2 = curX[1] - b2
  curX = (c1,c2)    # new X 
  sols.append((1-(curX[1]-3))**2+20*((curX[0]+3)-(curX[1]-3)**2)**2)
  for i in range(2):
    diff=abs(curX[i]-prevX[i])
    if diff<=precision:
      print("\nThe local minimum occurs at", curX)
      break
  print("Iteration",aIter+1,"\nX value is",curX) #Print iterations
  inters.append(aIter+1)
  
# Graph
plt.figure(figsize=(10, 3)) 
plt.plot(inters, sols)
plt.ylabel('f(x1,x2)')
plt.xlabel('Iteration')
plt.title("Gradient Descent")
plt.show()

```
